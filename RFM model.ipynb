# F 购买的频率 对订单进行count 
# M 订单金额 对订单金额进行sum

```python
rfm_data = sheet_all_new.groupby(['year','会员ID'],as_index=False).agg({"R": 'min',  # 计算最近一次订单时间
                                                   '订单号': 'count', # 计算订单频率
                                                   '订单金额': 'sum'})  # 计算订单总金额
rfm_data.sample(10)
```

```python
rfm_data.columns = ["year","会员ID","R","F","M"]
rfm_data.sample(5)
```

# 现在对R，F，M分段进行打分

```python
#要引用到pd.cut()
rfm_data.iloc[:,2:].describe().T #用iloc同时取几列
```

```python
#R的范围[-1,200,600,950,1460]
#F的范围[0,1,1,1,130]
#M的范围[0,69,188,1198,206253]
```

```python
R_bins = [-1,365,730,1095,1461]
F_bins = [0,1,2,3,130]
M_bins = [0,70,200,1200,206253]
rfm_data["R_score"] = pd.cut(rfm_data["R"],R_bins,labels= [i for i in range(len(R_bins)-1,0,-1)]) #从4到0倒排 ，时间间隔越久，打分越低
rfm_data["F_score"] = pd.cut(rfm_data["F"],F_bins,labels= [i+1 for i in range(len(F_bins)-1)]) #频率正排，频率越高，打分越高
rfm_data["M_score"] = pd.cut(rfm_data["M"],M_bins,labels= [i+1 for i in range(len(M_bins)-1)]) #数量越多，打分越高，正排
```

```python
rfm_data.sample(10)
rfm_data["会员ID"].drop_duplicates().count()
```

```python
# 上面i+1的目的是 为了R F M三者保持统一，最大都能为4，R的范围是【4，3，2，1】 F M的范围是【0，1，2，3】所以得+1才够得着R的范围
```
